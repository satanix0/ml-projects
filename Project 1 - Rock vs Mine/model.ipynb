### Binary Classification
<ol>
    <li> Train Test Distribution of data</li>
    <li> Applying Logistic Regression</li>
    <li> Verification of the Accuracy of the predictions based on previous data</li>
    <li> Returns either M or F for the new SONAR readings</li>
</ol>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Seperating data into input and target Variables
X = df.drop(columns=60,axis=1)
y = df[60]
```

```python
# Splits data inro training set and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

```python
# Model Training
model = LogisticRegression()

# Training the model with training data
model.fit(X_train, y_train)
```

```python
# Model Evaluation
# accuracy_score on training data

X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train,X_train_pred)
print('Training Data Accuracy : ', training_data_accuracy)
```

```python
# accuracy_score on test data

X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test,X_test_pred)
print('Test Data Accuracy : ', test_data_accuracy)
```

### Predictive System

```python
import random

# Generating 60 random values for input
start= 0.01
end = 0.5
num = 60 

input_data = []
# Set a length of the list to 60
for i in range(0, num):
    # any random float between start and end values
    # don't use round() if you need number as it is
    x = round(random.uniform(start, end), 3)
    input_data.append(x)

print(input_data)

inp = np.array(input_data)
inp_reshaped = inp.reshape(1,-1)

prediction= model.predict(inp_reshaped)

if prediction[0] == 'R':
    print('Phew! This Object is just a Rock')
else:
    print('Look Out! This Object is a Mine')  
```
